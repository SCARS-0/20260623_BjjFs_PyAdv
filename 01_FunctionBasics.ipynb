```python
# Map, Filter, Reduce

x = ["manish", "abhijeet", "rakesh", "vinayak", "shankar"]

# Map ############################################
resMap = list(map(str.capitalize, x))
print("\nresMap -->", resMap)

# Filter #########################################
def HasR(word):
    if('r' in str.lower(word)):
        return True
    else:
        return False

resFil = list(filter(HasR, x))
print("\nresFil -->", resFil)

# Reduce #########################################
def Add(a, b):
    return a + b

from functools import reduce
resRed = reduce(Add, x)
print("\nresRed -->", resRed)
```

```python
def Logger(fn):
    def wrapper(*vArgs, **kwArgs):
        print(f"{fn.__name__} called")
        ret = fn(*vArgs, **kwArgs)
        print(f"{fn.__name__} returned")
        return ret
    return wrapper

@Logger
def add(a, b):
    return a+b

@Logger
def mul(a, b):
    return a*b


print(add(1, 2))
print(mul(1, 2))
```

```python
def average_cost():
    """Coroutine: maintains the running average buy price as fills arrive."""
    total_cost = 0.0
    total_qty  = 0
    avg_price  = None
    try:
        while True:
            qty, price = yield avg_price        # receives via .send((qty, price))
            total_cost += qty * price
            total_qty  += qty
            avg_price   = round(total_cost / total_qty, 2)
    finally:
        print(f"[position closed] avg cost = {avg_price}, total qty = {total_qty}")


book = average_cost()
next(book)                        # PRIME: run up to the first yield (returns None)

print(book.send((100, 250.0)))    # buy 100 @ 250  -> 250.0
print(book.send((50,  280.0)))    # buy  50 @ 280  -> 260.0
print(book.send((150, 240.0)))    # buy 150 @ 240  -> 250.0

book.close()                      # triggers the finally block, then terminates
```

```python
# Fibonacci Generator
def FibGen(n):
    '''Generates a sequence of 'n' fibonacci numbers'''
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a+b

it = FibGen(10)
try:
    while(1):
        print(next(it))
except StopIteration:
    print("\nDone")
```

```python
class FibGen:
    def __init__(self, num):
        self.num = num

    def __iter__(self):
        self.a = 0
        self.b = 1
        self.count = 0
        return self
    
    def __next__(self):
        if self.count < self.num: 
            x = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return x
        else:
            raise StopIteration
        
myFib = FibGen(10)
myIter = iter(myFib)

for x in myIter:
    print(x, end="--")
print("Done")
```
