```python
# Positional Arg

def Add(a, b):
    return a + b

Add(1, 2)
```

```python
# Named Args / KeyWorded Args
def Add(a, b):
    return a + b

Add(b=2, a=1)
```

```python
def Add(a, b, c=0, d=0):
    return a + b + c + d

# Add(1, 2), Add(1, 2, 3)
Add(1, 2, 3, 4), Add(1, 2, 3)
```

```python
lst = [1, 2, 3, 4]
a, *others, d = lst
a, d, others
```

```python
lst = [1, 2, 3, 4]
a, *_, d = lst
a, d
```

```python
def Add(a, b, *others):     # Variable Args
    sum  = a + b
    print(others)
    for val in others:
        sum += val

    return sum


Add(1, 2), Add(1, 2, 3, 4), Add(1, 2, 3)
```
