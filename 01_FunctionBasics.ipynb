```python
def PrintName(ceo, cto, **others):      # Keyworded Variable list
    print(f"ceo = {ceo}")
    print(f"cto = {cto}")
    print(type(others), others)
    for position, person in others.items():
        print(f"{position} = {person}")


PrintName("Manish", "Abhijeet", cfo="Rakesh", cxo="Varun")
```

```python
def PrintEmp(id, name, city):
    print(f"{id = }")
    print(f"{name = }")
    print(f"{city = }")


def Logged_PrintEmp(*vArgs, **kwArgs):  # Universal Arg List
    print(f"vArg-count = {len(vArgs)}")
    print(f"kwArg-count = {len(kwArgs)}")

    print("Method called")
    PrintEmp(*vArgs, **kwArgs)
    print("Method returned")

Logged_PrintEmp(100, "Pravin", "Pune")
Logged_PrintEmp(100, city="Pune", name="Pravin")
```

```python
# Map, Filter, Reduce

x = ["manish", "abhijeet", "rakesh", "vinayak", "shankar"]

# Map ############################################
resMap = list(map(str.capitalize, x))
print("\nresMap -->", resMap)

# Filter #########################################
def HasR(word):
    if('r' in str.lower(word)):
        return True
    else:
        return False

resFil = list(filter(HasR, x))
print("\nresFil -->", resFil)

# Reduce #########################################
def Add(a, b):
    return a + b

from functools import reduce
resRed = reduce(Add, x)
print("\nresRed -->", resRed)
```
